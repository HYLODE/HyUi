```python
bed_fields = [
    "department", "room", "hl7_bed",
    "location_id", "location_string",
    "hl7_department", "hl7_room",
    "department_id", "room_id", "bed_id",
]
```

```python
beds[["bed_split_1", "bed_split_2"]] = beds["hl7_bed"].str.split("-", expand=True)
```

```python
beds
```

Now merge in departments to get 'floor' and then use floor to sort to create a floor_bed_index field

```python
departments = pd.read_json("department_defaults.json")
floors = departments.loc[:, ["department", "floor", "location_name"]]
```

```python
beds = beds.merge(floors, how="left", on="department")
```

```python
beds["bed_number"] = beds.loc[:,"bed_split_2"].str.extract('(\d+)')
beds.sort_values(["location_name", "floor", "department", "bed_number"], inplace=True, na_position="last")
```

!!dropping bed_index as I think this should be built dynamically in the application

use bed index and hand correct WMS floor 3 since this is the only floor that used '301' naming style for beds

perform the fillna with -1 _after_ the sort and the bed index is created so NaN values (those without bed numbers) go to the end; replace such numbers with NA

```python
beds["bed_number"] = beds["bed_number"].fillna(-1).astype(int)
# beds["bed_index"] = beds["bed_index"].fillna(-1).astype(int)
```

```python
beds.drop(columns=["bed_split_1", "bed_split_2"], inplace=True)
```

```python
beds.info()
```

```python
beds_defaults_icu = pd.read_json("bed_defaults_icu.json")
beds_defaults_icu.drop(columns=["department", "hl7_bed", "covid",], inplace=True)
beds_defaults_icu["closed"] = beds_defaults_icu["closed"].str.contains("True|1|TRUE")
beds_defaults_icu["xpos"] = beds_defaults_icu["xpos"].fillna(-1).astype(int)
beds_defaults_icu["ypos"] = beds_defaults_icu["ypos"].fillna(-1).astype(int)
beds_defaults_icu
```

```python
beds = beds.merge(beds_defaults_icu, on="location_string", how="left")
beds["closed"].fillna(False, inplace=True)
```

```python
beds.to_excel("bed_defaults.xlsx", index=False)
```
